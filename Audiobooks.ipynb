```python
import pandas as pd

# Function to clean 'Price' column by removing currency symbols and converting it to numeric
def clean_price_column(df, price_column='Price'):
    df[price_column] = df[price_column].replace('[₹,]', '', regex=True)
    df[price_column] = pd.to_numeric(df[price_column], errors='coerce')

# Unifying datasets based on common columns and cleaning the 'Price' column
def unify_datasets(audible_catlog, audible_catlog_adv, books_df, common_columns=['Book Name', 'Author', 'Price', 'Rating']):
    # Renaming 'Title' to 'Book Name' in books_df for consistency
    books_df.rename(columns={'Title': 'Book Name'}, inplace=True)
    
    # Selecting only the common columns
    audible_catlog_common = audible_catlog[common_columns]
    audible_catlog_adv_common = audible_catlog_adv[common_columns]
    books_df_common = books_df[common_columns]
    
    # Concatenating the datasets
    unified_df = pd.concat([audible_catlog_common, audible_catlog_adv_common, books_df_common], ignore_index=True)
    
    # Cleaning the 'Price' column
    clean_price_column(unified_df, 'Price')
    
    # Removing duplicates
    unified_df_cleaned = unified_df.drop_duplicates()
    
    return unified_df_cleaned

# Perform genre-based analysis after cleaning
def genre_based_analysis(df, group_column='Main Genre', price_column='Price', rating_column='Rating', rating_count_column='No. of People rated'):
    cleaned_df = df.groupby(group_column).agg({
        price_column: 'mean',
        rating_column: 'mean',
        rating_count_column: 'mean'
    }).reset_index()
    return cleaned_df

# Assuming the datasets are already loaded: audible_catlog, audible_catlog_adv, books_df
unified_df_cleaned = unify_datasets(audible_catlog, audible_catlog_adv, books_df)

# Displaying the cleaned and unified dataset
print("Unified Dataset:")
print(unified_df_cleaned.head())

# Performing genre-based analysis on the unified dataset
books_genre_grouped_cleaned = genre_based_analysis(books_df)

# Displaying the cleaned and grouped data
print("Genre-based analysis:")
print(books_genre_grouped_cleaned.head())
```
